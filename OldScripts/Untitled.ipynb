```python
import numpy as np
import pandas as pd
import csv
import datetime
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mvn
```

```python
# load all data into dataframe
def load_data(path, file_names, aliases):
    dates = {}
#     path = 'C:\\Users\\plant\\OneDrive\\Documents\\Neural Networks Project\\'
#     file_name_SP = 'SandPPrices'  
#     file_name_RE = 'RealEstateIndexDailySince2013'

    data_files = [file_name_SP, file_name_RE]
    aliases = ["sp", "re"]
    for data_set_idx in range(len(data_files)):
        cur_alias = aliases[data_set_idx]
        with open(path + data_files[data_set_idx] + '.csv', newline='') as csvfile:
            spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
            spamreader.__next__()
            for row in spamreader:
                try:
                    cur_date = datetime.datetime.strptime(row[0], '%m/%d/%Y')

                except:
                   continue
                if not cur_date in dates:
                    dates[cur_date] = {}
                # for each csv the alue you want might be in different rows
                if aliases[data_set_idx] == "sp":
                    dates[cur_date][cur_alias] = float(row[4])
                elif aliases[data_set_idx] == 're':
                    dates[cur_date][cur_alias] = float(row[1])

    frame = pd.DataFrame.from_dict(dates, orient='index')
    frame.columns = aliases
    frame.reset_index(inplace=True)
    frame = frame.rename(columns = {'index':'Date'})
    return frame
```

```python
# interval type is number of days so 30 days would be month etc
# intervals is the number of those intervals
# - interval to go backwards
def get_calculated_fields(frame, intervals, int_types):
    # calulate prev dates, futures dates
    # right now prev month, prev 2 wks, fur 2wks
    int_dict = { "dy": 1, "wk": 5, "yr": 252, "mn": 21}
    new_columns_dict = {}
    # instatiate new column names
    cols = frame.columns.values.tolist()
    for idx, interval in enumerate(intervals):
        for column in cols:
            if column != "Date":
                next_prev = "next"
                if interval < 0:
                    next_prev = "prev"
                new_columns_dict[column + "_" + next_prev + "_" + str(interval) + int_types[idx]] = []
                # like sp_prev_1mn
    for row_idx, _ in frame.iterrows():
        for inteval_idx, interval in enumerate(intervals):
            days = interval * int_dict[int_types[inteval_idx]]
            for column in frame:
                # for every column for every interval for every data point get percentage change
                if column != "Date":
                    pct_change = None
                    if row_idx + days >= 0 and row_idx + days < frame.shape[0]:
                        cur_value = frame.loc[row_idx].at[column]
                        interval_value = frame.loc[row_idx + days].at[column]
                        
                        # get percentage change
                        if days > 0: # looking into future
                            if cur_value == 0:
                                print("Should not be happening")
                                print(row_idx)
                            pct_change = (interval_value - cur_value) / cur_value
                        else: # looking into past
                            if interval_value == 0:
                                print("Should not be happening")
                                print(row_idx + days)
                            pct_change = (cur_value - interval_value) / interval_value
                    # add to correct list
                    next_prev = "next"
                    if interval < 0:
                        next_prev = "prev"
                    new_columns_dict[column + "_" + next_prev + "_" + str(interval) + int_types[inteval_idx]].append(pct_change)
                    
    # add calulated values to frame
    for key in new_columns_dict:
        frame[key] = new_columns_dict[key]
    return frame
                    
                        
                    
```
